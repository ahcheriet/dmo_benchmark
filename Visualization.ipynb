# Visualization of the Dynamic Multi-objective Benchmark Test Suite

Import libraries and python scripts which will be used to visualize the Pareto front and dynamic fitness landscape movement.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%pylab inline

# benchmark python script
import dynamic_benchmark
```

Now we'll create a function that will save an animation and embed it in an html string. 

```python
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=20, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

With this HTML function in place, we can use IPython's HTML display tools to create a function which will show the video inline:

```python
from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

Function used to calculate the value of the Pareto set. 

```python
""" Calculate the Pareto set solution
    
    INPUT:
        offset: 0 implies this is the global Pareto set.
        x1: float value to indicate the value of the first parameter
        t: time variable
        
    OUTPUT:
        output: Pareto optimal point (x1, ...)
"""
def pareto_set(offset, x1, t):
    output = [x1]
    output.extend([offset + np.sin(0.5*np.pi*(t-x1)) for e in range(20)])
    return output
```

```python
""" Efficient Pareto Filtering method. The code is obtained from: 
    http://code.activestate.com/recipes/578287-multidimensional-pareto-front
"""
def dominates(row, candidate_row):
    return sum([row[x] <= candidate_row[x] for x in range(len(row))]) == len(row)    


def pareto_filter(input_points, dominates=dominates):
    candidate_rowNr  = 0
    pareto_points    = set()
    dominated_points = set()
    while True:
        candidate_row = input_points[candidate_rowNr]
        input_points.remove(candidate_row)
        rowNr = 0
        non_dominated = True
        while len(input_points) != 0 and rowNr < len(input_points):
            row = input_points[rowNr]
            if dominates(candidate_row, row):
                # If it is worse on all features remove the row from the array
                input_points.remove(row)
                dominated_points.add(tuple(row))
            elif dominates(row, candidate_row):
                non_dominated = False
                dominated_points.add(tuple(candidate_row))
                rowNr += 1
            else:
                rowNr += 1

        if non_dominated:
            # add the non-dominated point to the Pareto frontier
            pareto_points.add(tuple(candidate_row))

        if len(input_points) == 0:
            break
    return pareto_points, dominated_points
```

### Plot Pareto set in the decision space

```python
from matplotlib import animation

# calculate the trajectory of the Pareto set
t_arr = [0.1* i for i in range(41)]
x1 = np.linspace(0, 1, 1000)
x_t = np.array([pareto_set(0, x1, t) for t in t_arr])

# set up figure & 3D axis for animation
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1], projection='3d')

# set the grid color and plane color
ax.grid(False)
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

# set up line
lines = ax.plot([], [], [], 'o', c='b')

# prepare the axes limits
ax.set_xlim((-0.2, 1.2))
ax.set_ylim((-1.2, 1.2))
ax.set_zlim((-1.2, 1.2))

# set point-of-view: specified by (altitude degrees, azimuth degrees)
ax.view_init(30, 135)
ax.patch.set_bounds(0,-0.2,1.0, 1.2)   # Set the bounds of the rectangle: left,bottom,width,height


# initialization function: plot the background of each frame
def init():
    for line in lines:
        line.set_data([], [])
    return lines


# animation function.  This will be called sequentially with the frame number
def animate(i):
    # we'll step two time-steps per frame.  This leads to nice results.
    for line in lines:
        line.set_data(x_t[i%40,0,:], x_t[i%40,1,:])
        line.set_3d_properties(x_t[i%40,2,:])
    fig.canvas.draw()
    return lines


# instantiate the animator.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=400, interval=10, blit=True)

# display animation
display_animation(anim)
```

### Plot the trade-off solution in the objective space

```python
# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(-0.2, 2), ylim=(-0.2, 2))
line, = ax.plot([], [], 'o')


# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

def animate(i):
    t = i % x_t.shape[0]
    x_arr = x_t[t, :, :]
    f_vec_list = list()
    f1_list, f2_list = list(), list()
    for j in range(x_arr.shape[1]):
        f_vec = dynamic_benchmark.DB8a(x_arr[:, j], 0.1*t)
        f_vec_list.append(f_vec)
        
    pareto_front, _ = pareto_filter(f_vec_list)
    for f1, f2 in pareto_front:
        f1_list.append(f1)
        f2_list.append(f2)
    line.set_data(f1_list, f2_list)
    return line,

from matplotlib import animation

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=400, interval=40, blit=True)

# display animation
display_animation(anim)
```

### Plot of the beta function (mixed modality, unimodal and multi-modal)

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation

# set up figure & 3D axis for animation
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1], projection='3d')
ax.grid(False)
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

# prepare the data
X_1 = np.arange(0, 1, 0.05)
X_2 = np.arange(-1, 1, 0.05)
X_1, X_2 = np.meshgrid(X_1, X_2)

# functions to compute the beta_mix
@np.vectorize
def vec_func_f1(x1, x2, t):
    x = [x1]+[x2]*20
    return dynamic_benchmark.beta_mix(x, t, dynamic_benchmark.g)[0]

@np.vectorize
def vec_func_f2(x1, x2, t):
    x = [x1]+[x2]*20
    return dynamic_benchmark.beta_mix(x, t, dynamic_benchmark.g)[1]

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This will be called sequentially with the frame number
def animate(i):
    ax.cla()
    ax.grid(False)
    ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
    ax.set_zlim((0, 20))
    Z = vec_func_f1(X_1, X_2, 0.1*i)
    surf = ax.plot_surface(X_1, X_2, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
        linewidth=0, antialiased=False)
    return surf

# instantiate the animator.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=400, interval=10, blit=True)

# save to gif file
#anim.save('beta_mix_3d.gif', writer='imagemagick', fps=4000, bitrate=50, dpi=200)

# display animation
print("Mixed modality beta function")
display_animation(anim)
```
